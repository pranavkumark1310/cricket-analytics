### Teams who won the toss and won the game

```python
def team_win_toss_game(team_name):
    res = {}
    count = 0
    for index,data in df3.iterrows():
        if data['toss_winner'] == team_name and data['winner'] == team_name:
            count += 1
            res.update({'team': team_name, 'wins': count})
    return res            
```

```python
def list_of_teams():
    teams = []
    for index,data in df3.iterrows():
        teams.append(data['team1'])
    
    teams = set(teams)
    teams = list(teams)
    return teams
```

```python
def win_toss_analysis():
    teams = list_of_teams()
    temp = []
    for team in teams:
        res = team_win_toss_game(team)
        temp.append(res)
    
    teams = []
    wins = []
    for data in temp:
        teams.append(data['team'])
        wins.append(data['wins'])
    
    figure(figsize=(12,10), dpi=80)
    plt.barh(teams, wins)
    plt.title('Number of matches won by teams batting first')
    plt.xlabel('Number of Matches')
    plt.ylabel('Teams')
    plt.show()

win_toss_analysis()
```

### Team win percentage at particular venue

```python
print(set(df3['venue']))
```

```python
def team_home_wins(team_name, venue_name):
    total_matches_at_venue = 0
    total_wins_at_venue = 0
    team_wins = {}
    for index,data in df3.iterrows():
        if data['venue'] == venue_name and (data['team1'] == team_name or data['team2'] == team_name):
            total_matches_at_venue += 1
    
    for index,data in df3.iterrows():
        if data['venue'] == venue_name and data['winner'] == team_name:
            total_wins_at_venue += 1
            win_percentage = (total_wins_at_venue / total_matches_at_venue) * 100
            team_wins.update({'team': team_name, 'venue': venue_name, 'wins': total_wins_at_venue, 'win_percentage': win_percentage})
    
    return team_wins    
```

```python
team_home_wins('Chennai Super Kings', 'M. Chinnaswamy Stadium')
```

```python
team_home_wins('Royal Challengers Bangalore', 'M. Chinnaswamy Stadium')
```

```python
team_home_wins('Chennai Super Kings', 'M.A. Chidambaram Stadium')
```

### Number of times teams who have won the toss

```python
def team_win_toss(team_name):
    res = {}
    count = 0
    for index,data in df3.iterrows():
        if data['toss_winner'] == team_name:
            count += 1
            res.update({'team': team_name, 'toss_win': count})
    return res
```

```python
def win_toss():
    teams = list_of_teams()
    temp = []
    for team in teams:
        res = team_win_toss(team)
        temp.append(res)
    
    teams = []
    wins = []
    for data in temp:
        teams.append(data['team'])
        wins.append(data['toss_win'])
    
    figure(figsize=(12,10), dpi=80)
    plt.barh(teams, wins)
    plt.title('Number of times winning the toss')
    plt.xlabel('Number of times winning the toss')
    plt.ylabel('Teams')
    plt.show()

win_toss()
```
